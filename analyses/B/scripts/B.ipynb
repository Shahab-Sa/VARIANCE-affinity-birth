```python
dfs['locus_v_call'] = dfs['v_call'].apply(lambda x: x.split('V')[0] if isinstance(x, str) and 'V' in x else x)
dfs['locus_c_call'] = dfs['c_call'].apply(lambda x: x[:3] if isinstance(x, str) else x)
```

```python
def determine_value(row):
    values = [row['locus'], row['locus_v_call'], row['locus_c_call']]
    non_nan_values = [v for v in values if not pd.isna(v)]
    
    if not non_nan_values:
        return 'NA'
    
    if len(non_nan_values) == 1:
        return non_nan_values[0]
    
    unique_values = set(non_nan_values)
    
    if len(unique_values) == 1:
        return unique_values.pop()
    else:
        return 'disagree'

# Apply the function to each row
dfs['locus_consensus'] = dfs.apply(determine_value, axis=1)
len(dfs[dfs['locus_consensus']=='disagree'])
```

```python
def ChainRadar(bcr, HeavyChains_operator, HeavyChains_threshold, LightChains_operator, LightChains_threshold): 
    # Needs to have a column named unique_bcs which has unique entries
    bcs_stats = pd.DataFrame() 
    grouping = bcr.groupby(by='unique_bcs', dropna=False)
    
    loci = ['IGH', 'IGK', 'IGL']
    for unique_bcs, df in grouping:
        for locus in loci:
            bcs_stats.loc[unique_bcs, '{}_seqs_num'.format(locus)] = len(list(df[df['locus'] == locus]['locus']))
            
    bcs_stats.reset_index(drop=False, inplace=True, names='unique_bcs')  

    bcs_stats['bc'] = 1
    bcs_stats['IGKL_seqs_num'] = bcs_stats['IGK_seqs_num'] + bcs_stats['IGL_seqs_num']
    print('bcs_stats:')
    display(bcs_stats)
    bcs_stats[['unique_bcs','IGH_seqs_num','IGK_seqs_num','IGL_seqs_num','IGKL_seqs_num']].to_csv(f"{sect_out_folder}/bcs_stats_10X_{'_'.join(set(bcr['dataset']))}.tsv", sep = '\t', index=False, header=True)
    
    grouping = bcs_stats[['IGH_seqs_num', 'IGKL_seqs_num', 'bc']].groupby(by=['IGH_seqs_num', 'IGKL_seqs_num'], dropna=False)['bc']
    IGH_IGKL_seqs_num = grouping.count().reset_index()
    IGH_IGKL_seqs_num = IGH_IGKL_seqs_num.astype(int).copy()
    print('IGH_IGKL_seqs_num:')
    display(IGH_IGKL_seqs_num)

    chainradarplot_df = pd.DataFrame(columns=list(set(IGH_IGKL_seqs_num['IGKL_seqs_num'])), 
                                    index=list(set(IGH_IGKL_seqs_num['IGH_seqs_num'])))

    # Populate the chain radar plot DataFrame
    for i in IGH_IGKL_seqs_num.index:
        chainradarplot_df.loc[IGH_IGKL_seqs_num.loc[i, 'IGH_seqs_num'], 
                              IGH_IGKL_seqs_num.loc[i, 'IGKL_seqs_num']] = IGH_IGKL_seqs_num.loc[i, 'bc']

    # Map string operators to actual functions
    dic_operators = {
        '<'  : operator.lt,
        '<=' : operator.le,
        '>'  : operator.gt,
        '>=' : operator.ge,
        '==' : operator.eq,
        '!=' : operator.ne,
    }
    
    # Apply the filtering conditions using the operators
    HeavyChains_condition = dic_operators[HeavyChains_operator](bcs_stats['IGH_seqs_num'], HeavyChains_threshold)
    LightChains_condition = dic_operators[LightChains_operator](bcs_stats['IGKL_seqs_num'], LightChains_threshold)
    
    bcs_stats_filtered = bcs_stats[HeavyChains_condition & LightChains_condition]

    filtered_df = bcr[bcr['unique_bcs'].isin(bcs_stats_filtered['unique_bcs'])]

    print('chainradarplot_df:')
    chainradarplot_df.rename(columns={'0': 'LC=0'}, inplace=True)
    display(chainradarplot_df)  # index is HC
    chainradarplot_df.to_csv(f"{sect_out_folder}/chainradarplot_10X_{'_'.join(set(bcr['dataset']))}.tsv", sep = '\t', index=True, header=True)
    
    return filtered_df

# Example usage
# Assuming `bcr` is your input DataFrame
# Replace '>' and '==' with your desired operators, and set the thresholds (e.g., 1.0).
```

```python
# Function to check if any of the values in a comma-separated v_call match
def match_v_calls(row, selection_v_call):
    v_calls_in_row = set(row.split(','))
    selection_v_calls = set(selection_v_call.split(','))
    
    # Return True if there's any intersection between the sets (i.e., any match)
    return bool(v_calls_in_row.intersection(selection_v_calls))
```

```python
dic_filtered_dfs_VGsub = {}
for dataset, df_to_subset in dic_chra_filtered_dfs.items():
    print(dataset, len(df_to_subset))

    # Filter the main DataFrame based on df_genes_usage_selection
    filtered_rows = []
    for _, sel_row in df_genes_usage_selection.iterrows():
        # Apply filtering based on status, locus, and mouse first, then match v_call values
        subset = df_to_subset[
            (df_to_subset['status'] == sel_row['status']) &
            (df_to_subset['locus'] == sel_row['locus']) &
            (df_to_subset['mouse'] == sel_row['mouse'])
        ]
        
        # Filter rows where v_call matches any of the comma-separated values in selection
        matches = subset['v_call'].apply(match_v_calls, selection_v_call=sel_row['v_call'])
        
        # Append matching rows
        filtered_rows.append(subset[matches])
    
    # Step 4: Concatenate all the filtered rows into the final subset DataFrame
    df_to_subset_subset = pd.concat(filtered_rows)
    
    # Display the result
    df_to_subset_subset

    group = ['status', 'locus', 'mouse', 'v_call']
    grouped = '_'.join(group)
    grouping = df_to_subset_subset.groupby(group, dropna=False)[['sequence']]
    grouping.count().reset_index().to_csv(f"{sect_out_folder}/6_filtered_Vgene_usages_{dataset}_after_chrad_filtering.tsv", sep = '\t', header=True, index=False)
    dic_filtered_dfs_VGsub[dataset] = df_to_subset_subset
    # let export after j_gene filtering
    #export_fasta(df_to_subset_subset)
    #df_to_subset_subset.to_csv(f"{sect_out_folder}/df_{dataset}.tsv", sep = '\t', header=True, index=False)
```
