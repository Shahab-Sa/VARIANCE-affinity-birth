```python
file = "../../D/data/output/dfs.tsv"
dfs=pd.read_csv(file, sep='\t',header=0, low_memory=False)
dfs
```

```python
set(dfs['frameshift'])
```

```python
#df['header'] = df['sequence_id'].apply(lambda x: f">{x}")
grouping = dfs.groupby('label', dropna=False)

for grouped, df_grouped in grouping:

    ref_row = pd.DataFrame({'sequence_id': [f"{grouped} ref"], 'new_sequence': [df_grouped.loc[:,'seq_reference'].values[0]]})


    df = pd.concat([ref_row, df_grouped[['sequence_id', 'new_sequence']]]).reset_index(drop=True)
    df['header'] = df['sequence_id'].apply(lambda x: f">{x}")
    
    df[['header', 'new_sequence']].stack().to_csv(f"{sect_out_folder}/{grouped}.fasta", header=False, index=False)
```
